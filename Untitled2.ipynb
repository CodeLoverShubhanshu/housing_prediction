```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load dataset
data_cal = fetch_california_housing()

# Extract features and target
X = data_cal.data
y = data_cal.target

# Create DataFrame
data = pd.DataFrame(X, columns=data_cal.feature_names)
data["SalePrice"] = y

data.head()
```

```python
plt.figure(figsize=(10,10))
cor = data.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.PuBu)
plt.show()
```

```python
cor_target = abs(cor["SalePrice"]) # absolute value of the correlation
relevant_features = cor_target[cor_target>-0.2] # highly correlated features
names = [index for index, value in relevant_features.items()] # getting the names of the features
names.remove('SalePrice') # removing target feature
print(len(names))# printing the features print (len(names) )
```

```python
from sklearn.model_selection import train_test_split 
x = data.drop("SalePrice", axis=1)
y = data["SalePrice"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print (x_train.shape)
print (x_test.shape) 
print (y_train.shape) 
print (y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
predictions = lr.predict(x_test)
print("Actual value of the house:- ", y_test[0])
print("Model Predicted Value:- ", predictions[0])
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,predictions)
rmse = np.sqrt(mse)
print(rmse)
```
